## Visualizing Earnings Based On College Majors

In this project we will be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. Using visualizations, we will be exploring questions & relationships between various columns from the dataset.

![college](img/college_pic.jpeg)

```python
#importing the libraries 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the dataset into a DataFrame and exploring the data
recent_grads = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv')

# exploring the initial enteries dataset
display(recent_grads.head())

# exploring the later enteries of the data
display(recent_grads.tail())

# exploring the columns
display(recent_grads.describe())
```

```python
# exploring the dimension of dataset
print("The data has got", recent_grads.shape[0], "rows and ",  recent_grads.shape[1], "columns")

# Exploring the NA's in the data
recent_grads.isnull().sum()

# dropping the rows with missing values
recent_grads.dropna(inplace = True)

# checking the dimension again
print("The data has got", recent_grads.shape[0], "rows and ",  recent_grads.shape[1], "columns")
```

```python
# scatter plot to explore the relationship between Sample size and median
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', color = "darkred");
```

```python
# scatter plot to explore the relationship between Sample size and Unemployement rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter');
```

```python
# scatter plot to explore the relationship between Full time (Number of employees employed 35 hours or more) and median
recent_grads.plot(x='Full_time', y='Median', kind='scatter', color = "green"); 
```

```python
# scatter plot to explore the relationship between Share Women (Women as share of total) and Unemployement rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', color = "black");
```

```python
# scatter plot to explore the relationship between Male graduates and Median
recent_grads.plot(x='Men', y='Median', kind='scatter', color = "brown");
```

```python
# scatter plot to explore the relationship between Female graduates and Median
recent_grads.plot(x='Women', y='Median', kind='scatter');
```

```python
""" 
Created individual scatter plots to visualize potential relationships between columns and histograms to visualize the 
distributions of individual columns. A scatter matrix plot combines both scatter plots and histograms into one grid of plots 
and allows us to explore potential relationships and distributions simultaneously.
"""

from pandas.plotting import scatter_matrix
```

```python
# 2 by 2 scatter matrix plot using the Sample_size and Median columns
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10));
```

```python
#3 by 3 scatter matrix plot using the Sample_size, Median, and Unemployment_rate columns
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10));
```
